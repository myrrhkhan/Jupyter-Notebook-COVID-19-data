```python
import pandas as plt
import matplotlib.pyplot as pd
```

```python
file = "Google Mobility_Report.csv"

df = plt.DataFrame()

plt.read_csv("Google Mobility_Report.csv")
```

```python
countries = ['United Arab Emirates', 'United States', 'Sweden', 'Bangladesh', 'Canada']

for country in countries:
    c = df[df['country_region'] == country]
    pd.plot(c['date'], c['retail_and_recreation_percent_change_from_baseline'], label = country)
    
pd.legend
pd.show
```
